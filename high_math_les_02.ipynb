## 3. Задание (в программе)

Напишите код на Python, реализующий построение графиков:
1.	окружности

```python
x = []
y1 = []
y2 = []
x0 = 0
y0 = 0
R = 10
for i in range((-1000 * R), (1000 * R + 1)):
    x_ = i/1000
    x.append(x_)
    y1.append(math.sqrt(R**2 - (x_ - x0)**2) + y0)
    y2.append(-math.sqrt(R**2 - (x_ - x0)**2) + y0)
plt.figure(figsize=(4, 4))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlim(-15, 15)
plt.ylim(-15, 15) 
plt.xlabel('x')
plt.ylabel('y');
```

```python
x = []
y1 = []
y2 = []
a = 10
b = 5
for i in range((-1000 * a), (1000 * a + 1)):
    x_ = i/1000
    x.append(x_)
    y1.append(b * math.sqrt(1 - x_**2/a**2))
    y2.append(-b * math.sqrt(1 - x_**2/a**2))
plt.figure(figsize=(4, 4))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlim(-12, 12)
plt.ylim(-10, 10) 
plt.xlabel('x')
plt.ylabel('y')
```

## 2. Задание 
Докажите, что при ортогональном преобразовании сохраняется расстояние между точками.

##### Доказательство:

Пусть L1 - расстояние м/у точкам в координатах x, y, а L2 - в координатах x', y', тогда:


$ L1^2 = (x1 - x2)^2 + (y1 - y2)^2; $

$ L2^2 = (x1' - x2')^2 + (y1' - y2')^2; $

$ x' = a11*x + a12*y + a13; $

$ y' = a21*x + a22*y + a23; $

Необходимо доказать, что:

$ L1 = L2 $

Подставляем преобразование координат в уравнение, раскрываем скобки, комбинируем переменные и выносим их за скобки

$ L2^2 = (a11*x1 + a12*y1 + a13 - a11*x2 - a12*y2 - a13)^2 + (a21*x1 + a22*y1 + a23 - a21*x2 - a22*y2 - a23)^2; $

$ L2^2 = (a11*(x1-x2) + a12*(y1-y2))^2 + (a21*(x1-x2) + a22*(y1-y2))^2; $

$ L2^2 = a11^2*(x1-x2)^2 + 2*a11*a12(x1-x2)*(y1-y2) + a12^2*(y1-y2)^2 + a21^2*(x1-x2)^2 + 2*a21*a22(x1-x2)*(y1-y2) + a22^2*(y1-y2)^2; $

$ L2^2 = (a11^2 + a21^2)*(x1 - x2)^2 + (a12^2 + a22^2)*(y1 - y2)^2 + 2*(x1-x2)*(y1-y2)*(a11*a12+a21*a22); $

По свойству ортогональных преобразований:

$ a11^2 + a21^2 = 1; $

$ a12^2 + a22^2 = 1; $

$ a11*a12 + a21*a22 = 0; $

Следовательно,

$ L2^2 = (x1 - x2)^2 + (y1 - y2)^2 = L1^2 $

Тогда,

$ L1 = L2 $

## 4. Задание (в программе)

Решите систему уравнений:

y = x2 – 1

exp(x) + x∙(1 – y) = 1


```python
import warnings
warnings.filterwarnings("ignore") 
from scipy.optimize import fsolve
```

```python
x = np.linspace(-5, 5, 501)
plt.grid(color="lightgrey")
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) - 1)/x + 1)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x*(1 - y) - 1)

x1, y1 = fsolve(equations, (-2, 5))
x2, y2 = fsolve(equations, (2, 8))
x3, y3 = fsolve(equations, (4, 20))
print(f'Решение 1: [{x1}, {y1}];\n'
      f'Решение 2: [{x2}, {y2}];\n'
      f'Решение 3: [{x3}, {y3}].')
```

##### Решите систему уравнений и неравенств:
y = x2 – 1

exp(x) + x∙(1 – y) > 1

```python
# Решением является кривая y = x2 - 1 в области значений y < (exp(x) - 1)/x + 1
x = np.linspace(-5, 5, 501)
plt.grid(color="lightgrey")
y = [(i**2 - 1) if (i**2 - 1) < ((np.exp(i) - 1)/i + 1) else None for i in x]
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
