### 4. Задание (в программе):

Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2.0*np.pi, 101)
k = 3
y = np.cos(x)
z = np.cos(k*x)

xnumbers = np.linspace(0, 7, 15)
ynumbers = np.linspace(-1, 1, 11)

plt.plot(x, y, 'r', x, z, 'g') # r, g - красный, зелёный цвета
plt.ylabel("Амплитуда")
plt.title("График функций cos(x) и cos(k∙x)")
plt.xticks(xnumbers)
plt.yticks(ynumbers)
plt.legend(['cos(x)', 'cos(k∙x)'])
plt.grid()
plt.axis([0, 6.5, -1.1, 1.1]) # [xstart, xend, ystart, yend]
plt.show()
```
