## Тема “Элементы теории вероятностей”

### 1. Задание (теорема сложения)
Найти вероятность выпадения 2 или 5 очков при подбрасывании игральной кости, на гранях которой имеются соответственно 1,2,3,4,5 и 6 очков. 

__Решение__:

Вероятность выпадения 2 очков: $P(2) = 1/6$<br>
Вероятность выпадения 5 очков: $P(5) = 1/6$<br>
Вероятность выпадения 2 или 5 очков $P(2,5) = P(2) + P(5) = 1/3 = 0.333(3)$

__Ответ__:

$P(2,5) = 0.333(3)$

### 2. Задание (теорема умножения)
Найти вероятность того, что при двух подбрасываниях той же самой игральной кости сначала выпадет 2, а затем 5.

__Решение__:

Обозначим вероятность того, что при двух подбрасываниях той же самой игральной кости сначала выпадет 2, а затем 5 как P(2,5), а вероятности выпадение 2 и 5 как P(2) и P(5) соответственно, тогда:

$P(2) = 1/6$<br>
$P(5) = 1/6$<br>

$P(2,5) = P(2) \cdot P(5) = \frac{1}{36} = 0.027(7)$

__Ответ__: 

$P(2,5) = 0.027(7)$

### 3. Задание 
Найти вероятность выпадения 2 и 5 очков при двух подбрасываниях той же самой игральной игральной кости. Обратите внимание на порядок выпадения костей!

__Решение__:
   
Вероятность выпадения 2 и 5 очков: $P(А) = 1/36$<br>
Вероятность выпадения 5 и 2 очков: $P(B) = 1/36$<br>

Наличие 2 и 5 в выпавших игральных костях может быть обеспечено 2 способами:
- сначала выпадет 2 потом 5
- сначала выпадет 5 потом 2

Тогда, вероятность выпадения 2 и 5 или 5 и 2 очков: $P(AB) = P(A)+P(B) = \frac{1}{36} + \frac{1}{36} = \frac{1}{18} = 0,055(5)$

__Ответ__: 

$P(AB) = 0,055(5)$


### 4. Задание  (Геометрическая вероятность +интервалы)
На отрезке AB длиной 20 см наугад отметили точку C. Какова вероятность, что она находится на расстоянии не более 9 см от точки А и не более 15 см от точки В?

__Решение__:

Чтобы соответствовать условиям задачи, точка C должна хаходится на участке отрезка AB от 5 до 9 см включительно. Таким образом, интервал значений отвечающим условиям задачи составит 4 см. Другими словами пропорция между удачными исходами и всеми возможными исходами будет равна 4/20.

$P(С) = 4/20 = 0.2$

__Ответ__: 

$P(С) = 0.2$

### 5. Задание. 
Телефонный номер состоит из 7 цифр. Какова вероятность, 
что это номер 8882227?

__Решение__:

Обозначим вероятность, что это номер 8882227 как P(A), тогда:

Есть 2 варианата P(A1) и P(A2)

- 1 и основной - номер не может начинаться с 0 
- 2 (его нельзя исключать, т.к. в условии не сказано об этом) - номер может начинаться с 0

Учитывая вышесказанное:

$P(A1) = \frac{1}{9*10^6} = 0.000000111$ - если номер не может начинаться с 0

$P(A2) = \frac{1}{10^7} = 0.0000001$ - если номер может начинаться с 0

__Ответ__: 

$P(A1) = 0.000000111$ - если номер не может начинаться с 0

$P(A2) = 0.0000001$ - если номер может начинаться с 0

### 6. Задание. 
Набирая номер телефона, абонент забыл 2 последние цифры, и, помня только то, что эти цифры различны и среди них нет нуля, стал набирать их наудачу. Сколько вариантов ему надо перебрать, чтобы наверняка найти нужный номер? Какова вероятность того, что он угадает номер с первого раза?

__Решение__:

Обозначим:
- A - количество вариантов которые надо пребрать.
- B  - событие, набраны две нужные цифры.
- P(B) - вероятность того, что он угадает номер с первого раза, тогда:

$А = 9*8 = 72$ - количество размещений из 9 цифр по 2

Эти исходы несовместны, равновозможны и образуют полную группу. Благоприятствует событию B лишь один исход. Искомая вероятность равна отношению числа исходов, благоприятствующих событию, к числу всех элементарных исходов:

$P(B) = \frac{1}{72} = 0.0139$

__Ответ__:

$А = 72$<br>
$P(B) = 0.0139$

### 7. Задание** (необязательное) 
Чёрный куб покрасили снаружи белой краской, затем разрезали на 27 одинаковых маленьких кубиков и как попало сложили из них большой куб. С какой вероятностью  все грани этого куба будут белыми?

__Решение__:

Все маленькие кубики разбиваются на 4 группы по количеству белых граней:
- 8 кубиков с 3 гранями, 
- 12 кубиков с 2 гранями, 
- 6 кубиков с 1 гранью,
- 1 черный кубик.

Если зафиксировать верхнюю грань, то получим 4 варианта расположения боковых граней.<br>
Существует 6 вариантов зафиксировать верхнюю грань и следовательно $6*4=24$ варианта расположения боковых граней при каждой фиксации.
Таким образом, существует $24$ способа повернуть маленький кубик.

Вероятность того, что 1-й кубик из 8 кубиков с 3 белыми гранями встанет на своё место равна

 $P_3^1=\frac{8}{27}\cdot\frac{3}{6}\cdot\frac{1}{4}=\frac{1}{27}\cdot\frac{3}{24}=\frac{1}{27}\cdot\frac{1}{8}$
 
Вероятность того, что 2-й кубик из 8 кубиков с 3 белыми гранями встанет на своё место, при том что 1- кубик уже на своём месте, равна

$P_3^2=\frac{7}{26}\cdot\frac{1}{8}$

    ...

Вероятность того, что 8-й кубик из 8 кубиков с 3 белыми гранями встанет на своё место (остальные 7 кубиков уже находятся на своих местах) равна

$P_3^8=\frac{1}{20}\cdot\frac{1}{8}$

Таким образом, вероятность того, что все кубики с 3-мя белыми гранями встанут на своё место

$P_3=\frac{1}{8^8}\frac{8\cdot7\cdot6\cdot5\cdot4\cdot3\cdot2\cdot1}{27\cdot26\cdot25\cdot24\cdot23\cdot22\cdot21\cdot20}=\frac{1}{8^8}\frac{8!}{(27!-19!)}$

Вероятность того, что 1-й кубик из 6 кубиков с 1-й белой гранью встанет на своё место, при том что все кубики с тремя белыми гранями находятся на своих местах, равна

$P_1^1=\frac{6}{19}\cdot\frac{1}{6}\cdot\frac{1}{4}=\frac{6}{19}\cdot\frac{1}{24}$
 
Вероятность того, что 2-й кубик из 6 кубиков с 1-й белой гранью встанет на своё место, при том что все кубики с тремя белыми гранями и первый кубик с одной белой гранью, находятся на своих местах равна

$P_1^2=\frac{5}{18}\cdot\frac{1}{24}$

    ...
    
Вероятность того, что 6-й кубик из 6 кубиков с 1-й белой гранью встанет на своё место равна

$P_1^6=\frac{1}{14}\cdot\frac{1}{24}$

Таким образом, вероятность того, что все кубики с 1-й белой гранью встанут на своё место, при том что все кубики с тремя белыми гранями находятся на своих местах

$P_1=\frac{1}{24^6}\frac{6\cdot5\cdot4\cdot3\cdot2\cdot1}{19\cdot18\cdot17\cdot16\cdot15\cdot14}=\frac{1}{24^6}\frac{6!}{(19!-13!)}$

Вероятность того, что 1-й кубик из 12 кубиков с 2-мя белыми гранями встанет на своё место, при том что все кубики с тремя и одной белыми гранями, находятся на своих местах, равна
    
$P_2^1=\frac{12}{13}\cdot\frac{2}{6}\cdot\frac{1}{4}=\frac{12}{13}\cdot\frac{1}{12}$
 
Вероятность того, что 2-й кубик из 12 кубиков с 2-мя белыми гранями встанет на своё место равна

$P_2^2=\frac{11}{12}\cdot\frac{1}{12}$

    ...
    
Вероятность того, что 12-й кубик из 12 кубиков с 2-мя белыми гранями встанет на своё место равна

$P_2^{12}=\frac{1}{2}\cdot\frac{1}{12}$

Таким образом, вероятность того, что все кубики с 2-мя белыми гранями встанут на своё место, при том что все кубики с тремя и одной белыми гранями, находятся на своих местах

$P_2=\frac{1}{12^{12}}\frac{12\cdot11\cdot10\cdot...\cdot2\cdot1\cdot2}{13\cdot12\cdot11\cdot...\cdot3\cdot2}=\frac{1}{12^{12}}\frac{12!}{(13!-1!)}$

Вероятность того, что кубик со всеми чёрными гранями встанет на своё место, при том что все остальные кубики уже находятся на своих местах, равна

$P_0^1=\frac{1}{1}\cdot\frac{24}{24}=1$

Полная вероятность того все грани этого куба будут белыми равна произведению всех найденных вероятностей

$P=P_3*P_2*P_1*P_0 = \frac{1}{8^8}\frac{8!}{(27!-19!)}\cdot\frac{1}{24^6}\cdot\frac{6!}{(19!-13!)}\cdot\frac{1}{12^{12}}\cdot\frac{12!}{(13!-1!)} = \frac{1}{8^8\cdot12^{12}\cdot24^6}\cdot\frac{8!\cdot6!\cdot12!}{(27!-19!)\cdot(19!-13!)\cdot(13!-1!)\cdot(1!)}=
\frac{1}{8^8\cdot12^{12}\cdot24^6}\cdot\frac{8!\cdot6!\cdot12!}{27!}
$

__Ответ__: 

Вероятность того, что все грани куба будут белыми равна:

$P=\frac{1}{8^8\cdot12^{12}\cdot24^6}\cdot\frac{6!\cdot8!\cdot12!}{27!}=\frac{6!\cdot8!\cdot12!}{8^8\cdot12^{12}\cdot24^6\cdot27!}$
